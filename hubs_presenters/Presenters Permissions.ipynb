# From Main Indico

```python
names = []
last_names = []
emails = []
with open('../registration_parserer/from_sam/for-marco-3075.csv', encoding = "ISO-8859-1") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        names.append(row[0])
        last_names.append(row[1])
        emails.append(row[2])
        line_count += 1
    print(f'Processed {line_count} lines.')
```

# From Poster Indico

```python
presenters = []
ids = []
with open('../posters_images/Nu2020 Posters - posters_200617_0930.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
#             print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            presenters.append(row[2])
            ids.append(row[0])
            line_count += 1
    print(f'Processed {line_count} lines.')
# presenters
# ids
```

```python

for p, pid in zip(presenters, ids):
    
    presenter = p.lower()
    
    found = False
    
    for n, l in zip(names, last_names):
        
        real_name = n.lower()
        real_last_name = l.lower()
        
        if real_name in presenter and real_last_name in presenter:
            print(f'found {pid}: {p} -> {n} {l}')
            found = True
            
    if not found:
        print(f'NOT found for {pid}: {p}')
            
```

```python
id_to_name = {}
for p, pid in zip(presenters, ids):
    
    presenter = [p]
    pid = int(pid)
    
    if pid in mapa:
        if mapa[pid] is not None:
            presenter = mapa[pid]
        else:
            continue
    
    for pres in presenter:
        found = False

        for n, l in zip(names, last_names):

            if n in pres and l in pres:
                print(f'found {pid}: {pres} -> {n} {l}')
                found = True
                id_to_name[pid] = f'{n} {l}'

        if not found:
            print(f'NOT found for {pid}: {pres}')
            id_to_name[pid] = f'{pres}'
            
```

```python
with open('../posters_images/poster_id_to_hubs_link_session4.json') as f:
    data = json.load(f)

# print(data)

id_to_hubsid = {}
for el in data:
    id_to_hubsid[el['poster_id']] = el['room_link'].replace('https://nu2020-hubs.org/', '')
    
id_to_hubsid
```
