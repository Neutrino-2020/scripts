```python
from os import listdir, system
from os.path import isfile, join
import json
import os
```

```python
with open('../indico_fetch/posters_200617_0930.json') as f:
    data = json.load(f)
```

```python
out_json = []

counter = 0
for entry in data['posters']:
#     print(entry['id'])
    if entry['poster_link'] == 'None':
        continue
        
    if entry['poster_session'] != 1:
        continue

    out_json.append({
        "poster_id": int(entry['id']),
        "room_number":int(counter/5)+1,
        "local_poster_number":counter%5+1
    })
    counter += 1
```

```python
out_json
```

```python
with open('posterid_to_room.json', 'w') as outfile:
    json.dump(out_json, outfile, indent=4)
```
